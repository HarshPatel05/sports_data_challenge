# NFL Combine Data Cleaning and Organization

This notebook focuses on one goal:

**Create one clean, organized combine dataset that can later be used in the master dataset.**

The notebook keeps the cleaning simple and conservative:
- load the raw file
- inspect the data
- clean column names
- clean basic text formatting
- check duplicates
- check missing values
- organize the final columns
- save one cleaned dataset


```python
import pandas as pd

# Show more columns when printing dataframes
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)
```

## 1. Load the raw dataset

```python
df = pd.read_csv("NFL_Combine_Since_2000.csv")

print("Rows, Columns:", df.shape)
df.head()
```

## 2. Inspect the raw data

This helps us understand what is in the file before making changes.


```python
print("Column names:")
print(df.columns.tolist())

print()
print("Data types:")
print(df.dtypes)
```

```python
print("First 10 rows:")
df.head(10)
```

## 3. Make a working copy and rename columns

We keep the raw dataframe unchanged and clean a copy.
The new column names are easier to read and easier to use later.


```python
clean_df = df.copy()

clean_df = clean_df.rename(columns={
    "Player": "player",
    "Position": "position",
    "School": "school",
    "Height": "height_in",
    "Weight": "weight_lb",
    "40-yd Dash": "forty_yd_dash",
    "Vertical Jump": "vertical_jump",
    "Bench Press": "bench_press",
    "Broad Jump": "broad_jump",
    "3-Cone Drill": "three_cone_drill",
    "20-yd Shuttle": "twenty_yd_shuttle",
    "Team": "team",
    "Round": "round",
    "Pick": "pick",
    "Year": "year"
})

clean_df.head()
```

## 4. Clean basic text formatting

This step only does light cleaning:
- remove leading and trailing spaces
- keep text columns consistent
- keep original football labels and meaning


```python
text_cols = ["player", "position", "school", "team"]

for col in text_cols:
    clean_df[col] = clean_df[col].astype("string").str.strip()

clean_df[["player", "position", "school", "team"]].head()
```

## 5. Check for duplicates

We check:
1. exact duplicate rows
2. duplicate records based on a more realistic combine key

For combine data, a safer key is:
- player
- year
- position
- school


```python
# Exact duplicate rows
exact_duplicates = clean_df.duplicated().sum()
print("Exact duplicate rows:", exact_duplicates)

# Duplicate records using a more realistic combine key
key_cols = ["player", "year", "position", "school"]
key_duplicates = clean_df.duplicated(subset=key_cols).sum()
print("Duplicate rows using player-year-position-school:", key_duplicates)
```

```python
# Show any duplicate records based on the combine key
duplicate_rows = clean_df[clean_df.duplicated(subset=key_cols, keep=False)].sort_values(key_cols)

duplicate_rows.head(20)
```

## 6. Check missing values

This tells us which columns are complete and which ones need attention later.
We are only inspecting missing values here, not filling them in.


```python
missing_summary = pd.DataFrame({
    "missing_count": clean_df.isna().sum(),
    "missing_percent": (clean_df.isna().mean() * 100).round(2)
}).sort_values("missing_percent", ascending=False)

missing_summary
```

## 7. Quick value checks

These checks help us catch obvious issues without overcomplicating the notebook.


```python
print("Year range:", clean_df["year"].min(), "to", clean_df["year"].max())
print("Unique positions:", clean_df["position"].nunique())
print("Unique schools:", clean_df["school"].nunique())
```

```python
numeric_cols = [
    "height_in", "weight_lb", "forty_yd_dash", "vertical_jump",
    "bench_press", "broad_jump", "three_cone_drill", "twenty_yd_shuttle",
    "round", "pick", "year"
]

clean_df[numeric_cols].describe().T
```

## 8. Organize the final cleaned dataset

We keep the columns in a clean order so the dataset is easier to read and easier to merge later.
We are not creating extra derived columns here.


```python
final_columns = [
    "player",
    "year",
    "position",
    "school",
    "height_in",
    "weight_lb",
    "forty_yd_dash",
    "vertical_jump",
    "bench_press",
    "broad_jump",
    "three_cone_drill",
    "twenty_yd_shuttle",
    "team",
    "round",
    "pick"
]

combine_clean = clean_df[final_columns].copy()
combine_clean.head()
```

## 9. Final review

A last look at the cleaned table before saving it.


```python
print("Final shape:", combine_clean.shape)
combine_clean.info()
```

```python
combine_clean.head(10)
```

## 10. Save the cleaned dataset

This is the main output from the notebook.


```python
combine_clean.to_csv("cleaned_combine_Data.csv", index=False)

print("Saved file: cleaned_combine_Data.csv")
```

## Final note

This notebook intentionally keeps the cleaning light and readable.

It does **not**:
- create extra summary CSV files
- create derived columns like BMI
- group draft rounds into buckets
- remap real position labels into other positions

It focuses only on producing **one clean combine dataset** that can later be used in the master merge.
